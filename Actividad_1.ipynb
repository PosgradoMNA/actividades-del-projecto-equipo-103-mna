# **1. Conceptos de almacenamiento y recuperación de información**

## **1.1 Fundamentos de bases de datos y para ciencia de datos**

Un elemento que destaca en la presentación son los 3 gráficos que muestran 3 conceptos comúnmente confundidos: Data Analyst , Data science & data engineer. Claramente se puede observar de ese gráfico las habilidades de cada una de estas profesiones pero algo que notoriamente tienen en común es el uso de bases de datos.

Un científico de datos dedica aproximadamente 80% de su tiempo en la recopilación y limpieza de datos. Este necesitará diseñar, consultar y/o interactuar con bases de datos. 

Una base de datos es una recopilación organizada de información o datos estructurados que se almacenan de forma electrónica en un sistema. También existen bases de datos no relacionadas ( NoSQL database) , están orientadas a almacenar documentos,es decir, se puede recuperar datos en formatos que no sean tablas (ej.MongoDB, Elasticsearch). 

Algunos ejemplos de sistemas de bases de datos estructuradas son: MySQL, Microsfot SQL server, IBM Db2, etc. 

## **1.2 Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.**

La diferencia entre una base de datos y un data warehouse es que este último es un conjunto de múltiples bases de datos relacionales con el que se puede analizar información, es decir, una base de datos registra mientras que el otro agrega datos para análisis. El principal propósito de este es proveer información a través de la inteligencia de negocio (BI).

Algunos ejemplos de data warehouse :  Amazon Redshift, Snowflake,y Google BigQuery, etc.  

De estos sistemas se puede obtener información pero la mayoría de las veces tiene que pasar por un proceso de limpieza antes de poder presentarlos a modo resumen ejecutivo en dashboards inteligentes ( Tableu, PowerBI, etc.)

## 2.3 Sustitución de datos

Creamos una copia del DataFrame para no manipular los datos originales y poder comparar.

```python
ndf = df.copy()
```

Reemplazamos los valores numericos con el promedio.

```python
ndf["EDAD"].fillna(ndf["EDAD"].mean(), inplace=True)
ndf["ATRASO_SEP2005"].fillna(ndf["ATRASO_SEP2005"].mean(), inplace=True)
ndf["ATRASO_AUG2005"].fillna(ndf["ATRASO_AUG2005"].mean(), inplace=True)
ndf["ATRASO_JUL2005"].fillna(ndf["ATRASO_JUL2005"].mean(), inplace=True)
ndf["ATRASO_JUN2005"].fillna(ndf["ATRASO_JUN2005"].mean(), inplace=True)
ndf["ATRASO_MAY2005"].fillna(ndf["ATRASO_MAY2005"].mean(), inplace=True)
ndf["ATRASO_APR2005"].fillna(ndf["ATRASO_APR2005"].mean(), inplace=True)
ndf["EDO CTA_SEP2005"].fillna(ndf["EDO CTA_SEP2005"].mean(), inplace=True)
ndf["EDO CTA_AUG2005"].fillna(ndf["EDO CTA_AUG2005"].mean(), inplace=True)
ndf["EDO CTA_JUL2005"].fillna(ndf["EDO CTA_JUL2005"].mean(), inplace=True)
ndf["EDO CTA_JUN2005"].fillna(ndf["EDO CTA_JUN2005"].mean(), inplace=True)
ndf["EDO CTA_MAY2005"].fillna(ndf["EDO CTA_MAY2005"].mean(), inplace=True)
ndf["EDO CTA_APR2005"].fillna(ndf["EDO CTA_APR2005"].mean(), inplace=True)
ndf["PAGO_SEP2005"].fillna(ndf["PAGO_SEP2005"].mean(), inplace=True)
ndf["PAGO_AUG2005"].fillna(ndf["PAGO_AUG2005"].mean(), inplace=True)
ndf["PAGO_JUL2005"].fillna(ndf["PAGO_JUL2005"].mean(), inplace=True)
ndf["PAGO_JUN2005"].fillna(ndf["PAGO_JUN2005"].mean(), inplace=True)
ndf["PAGO_MAY2005"].fillna(ndf["PAGO_MAY2005"].mean(), inplace=True)
ndf["PAGO_APR2005"].fillna(ndf["PAGO_APR2005"].mean(), inplace=True)
```

Reemplazamos los valores categoricos con la moda.

```python
ndf["SEXO"].fillna(ndf["SEXO"].mode()[0], inplace=True)
ndf["ESCOLARIDAD"].fillna(ndf["ESCOLARIDAD"].mode()[0], inplace=True)
ndf["EDO_CIVIL"].fillna(ndf["EDO_CIVIL"].mode()[0], inplace=True)
ndf["Y_N Class"].fillna(ndf["Y_N Class"].mode()[0], inplace=True)
```
